What is Flask Framework? What are the advantages of Flask Framework?

Ans: Flask is a micro web framework written in Python that allows developers to build web applications quickly and easily.

1.Easy to Use: Flask has a simple and intuitive interface, making it easy to learn and use. It provides a lightweight web application development experience and is suitable for small to medium-sized projects.

2.Flexible: Flask is highly customizable and allows developers to add new functionalities and features as per the requirement. It provides a wide range of extensions and libraries to extend the functionality of the application.

3.Lightweight: Flask is a lightweight framework with minimal dependencies, making it easy to install and use. It does not require complex configurations and setup, making it a suitable choice for small to medium-sized applications.

4.Scalable: Flask provides scalability and can be used to build large and complex web applications. It is a highly modular and extensible framework that can be used to build complex applications with ease.

5.Integrated Development: Flask provides integrated development features, such as a built-in debugger, testing tools, and development server, which makes development and debugging easier.

6.Large Community: Flask has a large and active community of developers who contribute to the development and maintenance of the framework. This community provides support, tutorials, and extensions, making it easier to learn and use the framework.

What is App routing in Flask? Why do we use app routes?

Ans : In Flask, app routing refers to the process of mapping URL patterns to Python functions, which are defined as routes within a Flask application.

App routing is achieved in Flask using the @app.route decorator. This decorator is used to define a new route by specifying the URL pattern and the HTTP methods that are allowed for that route.

## 5 Question
What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans: 

The url_for() function is used in Flask for URL building. It generates a URL for the specified endpoint and accepts one or more arguments to build the URL dynamically. It takes the endpoint name as the first argument and any number of keyword arguments as additional parameters to build a dynamic URL based on the named routes defined in the Flask application.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def profile(username):
    return f'User: {username}'

@app.route('/')
def index():
    return f'The URL for profile page is: {url_for("profile", username="sagar panchal")}'
if __name__=="__main__":
    app.run(host="0.0.0.0")
