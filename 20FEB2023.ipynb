## 1 Question

Explain GET and POST methods.

Ans: GET and POST are two HTTP request methods used for retrieving and sending data between a client (such as a web browser) and a server.

GET Method:
The GET method is used to request data from a specified resource, typically to retrieve a web page or some other data from a server. When a GET request is made, the client sends a request to the server with parameters (if any) attached to the URL. The server then responds with the requested data, usually in the form of an HTML page or some other resource that can be displayed in a web browser. The GET method is generally considered "safe" because it does not modify any data on the server, and the requested data is visible in the URL.

POST Method:
The POST method is used to send data to a server to create or update a resource. When a POST request is made, the client sends data in the request body to the server. This data is typically sent using a web form, but can also be sent programmatically. The server then processes the data and creates or updates the specified resource. The POST method is generally considered "unsafe" because it can modify data on the server, and the data sent in the request body is not visible in the URL.

In summary, the main difference between GET and POST is that GET is used to retrieve data from a server while POST is used to send data to a server to create or update a resource. GET requests are generally considered "safe" because they do not modify any data on the server, while POST requests are considered "unsafe" because they can modify data on the server.

## 2 Question

Why is request used in Flask?

Ans : In Flask, request is a module that allows you to access incoming request data in a Flask application.

When a client sends a request to a Flask application, the application creates a request object that contains all of the information about the request, including any data that was submitted with the request. This can include things like form data, JSON data, query parameters, and more.

The request object provides several methods and attributes that allow you to access this data, such as:

request.method: Returns the HTTP method used in the request (e.g. GET, POST, PUT, DELETE).
request.form: Returns a dictionary containing any submitted form data.
request.args: Returns a dictionary containing any submitted query string parameters.
request.json: Returns the parsed JSON data submitted with the request.
request.files: Returns a dictionary containing any submitted files.

## 3 question

Why is redirect() used in Flask?

Ans : In Flask, redirect() is a function that allows you to redirect a user's browser to a different URL.

Redirects are often used in web applications to send users to different pages or routes based on some condition or action. For example, after a user successfully logs in to a Flask application, the application might redirect the user to their profile page.

The redirect() function takes a single argument, which is the URL that you want to redirect the user to. For example, redirect('/profile') would redirect the user to the /profile route in your Flask application.

## 4 Question

What are templates in Flask? Why is the render_template() function used?

Ans : In Flask, templates are files that define the structure and layout of your application's HTML pages. Templates allow you to separate your application's logic from its presentation, making it easier to maintain and modify your application over time.

Templates are typically written in HTML, and can include placeholders for dynamic content that will be generated by your Flask application. For example, you might create a template for a user's profile page that includes placeholders for the user's name, email address, and other information that will be loaded dynamically from the application's database.

To render a template in Flask, you use the render_template() function. This function takes the name of the template file as its first argument, and any additional keyword arguments that you want to pass to the template as variables.
